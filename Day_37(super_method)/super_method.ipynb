```python
class Car:
    def __init__(self,type):
        self.type = type
    def start(self):
        print("Car Start...")
    def stop(self):
        print("Car Stop...")
class BmwCar(Car):
    def __init__(self,name,type):
        self.name = name
        super().__init__(type)
        super().start()
car1 = BmwCar("TATA","Electric")
print(car1.type)
```
