```python
#Defult Constructor
class Student:
    name = "Shubham Jana"
    roll = 34
s1 = Student
print(s1.name)
```

```python
#Parameterised Constructor.
class Student:
    def __init__ (self,name,roll):
        self.name = name
        self.roll = roll
s2 = Student("Rajiv", 78)
print(f"The student name is {s2.name} . ")
print(f"Their Roll number:- {s1.roll} . ")
```
