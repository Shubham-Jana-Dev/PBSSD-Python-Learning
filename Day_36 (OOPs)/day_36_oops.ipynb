```python
#Parameterised Constructor.
class Student:
    def __init__ (self,name,roll):
        self.name = name
        self.roll = roll
s2 = Student("Rajiv", 78)
print(f"The student name is {s2.name} . ")
print(f"Their Roll number:- {s1.roll} . ")
```

```python
class Computer:
    def config(self):
        print("i5, 16gb, 1TB")
com1 = Computer()
com1.config()
```

```python
class Computer:
    def __init__(self,config):
        self.config = config
com1 = Computer("M1 8gb 256gh")
print(com1.config)
```

```python
class System:
    def __init__(self):
        print("Hey Shubham here is your system :)")
    def config(self):
        print("M1 8GB 2025")
emp1 = System()
emp2 = System()
print(emp1.config)
print(emp2.config)
```

```python
class System:
    def __init__(self,cpu,ram):
        self.cpu = cpu
        self.ram = ram
        print("Your Computer's specifications are: -")
        print(f"CPU = {self.cpu}\nRAM = {self.ram}")
emp1 = System("Intel i5 12th gen","16GB")
```

```python
'''
​Create a class called Scholar that captures the transition from a student's current reality to their aspirations.
​Your Task:
​The Initialization: Write an __init__ function that takes three parameters: name, current_location, and dream_destination.
​The "Vibe" Attribute: Inside __init__, create a fourth attribute called state_of_mind. This shouldn't be a parameter—set it to "Processing..." by default.
​The Method: Create a method called sync_reality().
​If the current_location is the same as the dream_destination, set state_of_mind to "At Peace".
​Otherwise, set it to "Rage Coding".
​The Output: Create an instance of Scholar for yourself and print:
"[Name] is currently in [Location], dreaming of [Destination]. Status: [Status]."
'''

class Scholar:
    def __init__(self,name,current_location,dream_destination):
        self.name = name
        self.current_location = current_location
        self.dream_destination = dream_destination
        self.state_of_mind  = "Processing..."
        self.sync_reality()
    def sync_reality(self):
        if self.current_location == self.dream_destination:
            self.state_of_mind = "At Peace"
        else:
            self.state_of_mind = "Rage Coding"
        print(f"{self.name} is currently in {self.current_location}, dreaming of {self.dream_destination}. Status: {self.state_of_mind}")
s1 = Scholar("Shubham","exhausted by the internship and training seven days in a week, 12 hour and university pressure","lead software engineer")
```
