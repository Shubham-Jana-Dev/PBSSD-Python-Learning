```python
#Defult Constructor
class Student:
    name = "Shubham Jana"
    roll = 34
s1 = Student
print(s1.name)
```

```python
#Parameterised Constructor.
class Student:
    def __init__ (self,name,roll):
        self.name = name
        self.roll = roll
s2 = Student("Rajiv", 78)
print(f"The student name is {s2.name} . ")
print(f"Their Roll number:- {s1.roll} . ")
```

```python
class Computer:
    def config(self):
        print("i5, 16gb, 1TB")
com1 = Computer()
com1.config()
```

```python
class Computer:
    def __init__(self,config):
        self.config = config
com1 = Computer("M1 8gb 256gh")
print(com1.config)
```

```python
class System:
    def __init__(self):
        print("Hey Shubham here is your system :)")
    def config(self):
        print("M1 8GB 2025")
emp1 = System()
emp2 = System()
print(emp1.config)
print(emp2.config)
```

```python
class System:
    def __init__(self,cpu,ram):
        self.cpu = cpu
        self.ram = ram
        print("Your Computer's specifications are: -")
        print(f"CPU = {self.cpu}\nRAM = {self.ram}")
emp1 = System("Intel i5 12th gen","16GB")
```
