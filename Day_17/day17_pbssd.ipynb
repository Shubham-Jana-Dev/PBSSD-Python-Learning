```python
"""
Write a function called reverser that uses *args.
It should accept any number of strings (e.g., "Python", "Web", "PBSSD").
Use a list comprehension inside to reverse each string.
Return the list of reversed strings.
"""
def reversed_string(*args):
    return [s[::-1] for s in args]
original_strings = ("Python", "Web", "PBSSD")
result = reversed_string(*original_strings)
print(f"Reversed string: {result}")
```

```python
def build_profile(name, **info):
    """
    'name' is required. 
    '**info' captures any other named details.
    """
    print(f"User: {name}")
    print(f"Additional Data: {info}")

# We pass data as key=value pairs
build_profile("Souvik", age=21, city="Kolkata", role="Python Dev")
```

```python
'''
New Task (The "Redemption" Challenge)
Write a function called math_master.
The first argument should be a multiplier (a number).
The *args should collect several other numbers.
The function should return a list where each number in args is multiplied by the multiplier.
Example: math_master(2, 5, 10, 15) should return [10, 20, 30].
'''
def math_master(multiplier, *args):
    return [n * multiplier for n in args]
result = math_master(2,5,10,15)
result
```

```python
"""
The "Combined" Challenge
Write a function called super_logger.
It should take one required argument: event_name.
It should take *args to collect any tags (like "High-Priority", "Server-1").
It should take **kwargs to collect details (like user="admin", status="success").
Inside the function, just print all three.

Example Call: super_logger("Login", "Alert", "Security", user="Srijan", ip="192.168.1.1")
"""
def super_logger(event_name,*args,**kwargs):
    print(f"User: {event_name}\nRole: {args}\nAdditional Data: {kwargs}")
    return event_name, args,kwargs
super_logger("Shubham Jana","Security",user="admin",ip="192.168.1.1")
```
