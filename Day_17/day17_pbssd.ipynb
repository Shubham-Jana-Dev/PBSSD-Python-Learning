```python
def squre_number():
    num = int(input("Enter the number: "))
    x = lambda n : n*n
    print(f"The squre of the {num} is {x(num)}")
# squre_number()
def apply_operation(numbers, func):
    return[func(n) for n in numbers]
my_nums = [1,2,3,4,5,6]
squres = apply_operation(my_nums,lambda n : n**2)
cubes = apply_operation(my_nums, lambda n : n**3)
print("Squres : ",squres)
print("Cubes : ",cubes)
```

```python
"""
Write a function called reverser that uses *args.
It should accept any number of strings (e.g., "Python", "Web", "PBSSD").
Use a list comprehension inside to reverse each string.
Return the list of reversed strings.
"""
def reversed_string(*args):
    return [s[::-1] for s in args]
original_strings = ("Python", "Web", "PBSSD")
result = reversed_string(*original_strings)
print(f"Reversed string: {result}")
```

```python
def build_profile(name, **info):
    """
    'name' is required. 
    '**info' captures any other named details.
    """
    print(f"User: {name}")
    print(f"Additional Data: {info}")

# We pass data as key=value pairs
build_profile("Souvik", age=21, city="Kolkata", role="Python Dev")
```
