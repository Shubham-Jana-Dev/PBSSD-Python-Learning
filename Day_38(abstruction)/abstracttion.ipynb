*1.* Abstruction class is a class that contains one or more abstraction method is called **Abstract class**.

```python
from abc import ABC, abstractmethod
class car(ABC):
    def show(self):
        print("Every can has wheels.")
    @abstractmethod
    def speed(self):
        pass
class Maruti(car):
            def speed(self):
                print("speed is 80 km/h")
class suzuki(car):
    def speed(self):
         print("Speed is 60 km/h")
obj1 = Maruti()
obj1.show()
obj1.speed()

obj2 = suzuki()
obj2.show()
obj2.speed()
```

```python
from abc import ABC, abstractmethod
class Vechicle(ABC):
    @abstractmethod
    def move(self):
        pass
class Car(Vechicle):
    def move(self):
        print("Car drived on rode...")
class Boat(Vechicle):
    def move(self):
        print("Boat sail in water...")
vechicle1 = Car()
vechicle2 = Boat()
vechicle1.move()
vechicle2.move()
```
